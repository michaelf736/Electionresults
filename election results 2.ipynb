# Webscraping the UK Election Results

```python
#Import libraries
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
#Get the webpage
res = requests.get("http://www.bbc.co.uk/news/election/2017/results")
```

```python
#Parse using Beautiful Soup
soup = BeautifulSoup(res.content, 'html.parser')
```

```python
#Find the correct table (3rd on the page)
table = soup.find_all('table')[2]
```

```python
#Construct pandas data frame from the strings in the table
df = pd.read_html(str(table))
```

```python
#Format
df = df[0]
```

```python
df
```

```python
#Trim the unnecessary words
df['Party'] = df['Party'].map(lambda x: str(x)[5:])
df['Net change in seats +/-'] = df['Net change in seats +/-'].map(lambda x: str(x)[20:])
df['Votes'] = df['Votes'].map(lambda x: str(x)[5:])
df['Vote Share'] = df['Vote Share'].map(lambda x: str(x)[10:])
df['Net percentage change in seats +/- %'] = df['Net percentage change in seats +/- %'].map(lambda x: str(x)[19:])
```

```python
#drop the last row
df.drop(df.index[23])
```
